```python
ACCA_Websites = []

for i in ACCA_Firms:
    ACCA_Websites.append(i[6])

def orderedUniques(seq, idfun=None): 
    # order preserving
    if idfun is None:
        def idfun(x): return x
    seen = {}
    result = []
    for item in seq:
        marker = idfun(item)
        # in old Python versions:
        # if seen.has_key(marker)
        # but in new ones:
        if marker in seen: continue
        seen[marker] = 1
        result.append(item)
    return result

ACCA_Websites = orderedUniques(filter(None, ACCA_Websites))
```

```python
import scrapy
import logging
from scrapy.crawler import CrawlerProcess
from scrapy.spiders import CrawlSpider, Rule
from scrapy.linkextractors import LinkExtractor
from urlparse import urlparse

class MySpider(CrawlSpider):
    name = "ACCA Crawler"

# Use this code to supress logging for long crawls
    def __init__(self, category=None, *args, **kwargs):
        super(MySpider, self).__init__(*args, **kwargs)
        logging.getLogger("scrapy").setLevel(logging.INFO)
    
    custom_settings = {
        # Relatively slow crawl. Probably won't get IP banned.
        'DOWNLOAD_DELAY': 6,
        # High level of concurrency, but this is across all domains, not per domain.
        # However, too much concurrency leads to timeouts on our side.
        'CONCURRENT_REQUESTS':150,
        # Not a very deept crawl, otherwise it gets silly.
        'DEPTH_LIMIT':1,
        # These make this a "breadth first" crawl.
        'DEPTH_PRIORITY':1,
        'SCHEDULER_DISK_QUEUE':'scrapy.squeues.PickleFifoDiskQueue',
        'SCHEDULER_MEMORY_QUEUE':'scrapy.squeues.FifoMemoryQueue'
        # Use this code to be kinder to target websites
        #,'ROBOTSTXT_OBEY':True
        #,'AUTOTHROTTLE_ENABLED':True
    }

    allowed_domains = []
    start_urls = []    

    for i in ACCA_Websites[1300:1400]:
        allowed_domains.append(i.replace("https://","").replace("http://",""))
        allowed_domains.append(i.replace("https://www.","").replace("http://www.",""))
        start_urls.append(i)

    rules = (Rule(LinkExtractor(allow=(),),callback="parse_item",follow=True),)    

    def parse_start_url(self, response):
        print(response.url)
        parsed_uri = urlparse(response.url)
        domain = '{uri.netloc}'.format(uri=parsed_uri)

        ACCA_website_data.append([domain.replace("www.",""),response.url,response.meta['depth'],response.body])
        
    def parse_item(self, response):
        print(response.url)
        parsed_uri = urlparse(response.url)
        domain = '{uri.netloc}'.format(uri=parsed_uri)

        ACCA_website_data.append([domain.replace("www.",""),response.url,response.meta['depth'],response.body])
```

```python
ACCA_website_data = []

process = CrawlerProcess()
process.crawl(MySpider)
process.start()
```
